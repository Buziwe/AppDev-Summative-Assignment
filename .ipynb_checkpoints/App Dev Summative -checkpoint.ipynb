### Defining predictors and response variables.

```python
# Values of target/predictors variable
y = data2['Power Generated in MW'].values

# Values of attributes/response variable
dataset = data2.drop(['Power Generated in MW', 'Day', 'Month'], axis=1) #drop colunms predictir column and outlier detection column for modelfitting
X = dataset.values
```
